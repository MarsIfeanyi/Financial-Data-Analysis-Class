```python
import pandas as pd
```

```python
class FinancialDataAnalysis():

    def __init__(self):
        self.get_data()
            
    def get_data(self):
        ''' read and retrieves the data from the excel file
        '''
        
        rawData = pd.read_excel("SP500.xls", parse_dates= ["Date"], index_col = "Date").Close.to_frame()
        rawData.rename(columns = {"Close":"price"}, inplace = True)
        self.data = rawData 
        return self.data
```

```python
stock = FinancialDataAnalysis()
```
